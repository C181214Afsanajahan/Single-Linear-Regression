```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20.0, 10.0)
```

```python
data = pd.read_csv('year_population.csv')
print(data.shape)
data.head()
```

```python
X = data['year'].values
Y = data['population'].values
```

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)
```

```python
n = len(X)
```

```python
# Formula : m = [(x - x_bar)* (y - y_bar)] / [ (x - x_bar)^2 ]
# y_bar = (m * x_bar) + c 
# From above equation, we can calculate the value of c as:
# c = y_bar - (m * x_bar)
# where x_bar = mean_x, and y_bar = mean_y
```

```python
numerator = 0
denominator = 0
for i in range(n):
    numerator   += (X[i] - mean_x) * (Y[i] - mean_y)
    denominator += (X[i] - mean_x) ** 2
m = numerator / denominator
c = mean_y - (m * mean_x)
```

```python
print('Value of m: ',m)
print('Value of c: ',c)
```

```python
max_x = np.max(X) + 100
min_x = np.min(X) - 100
```

```python
x = np.linspace(min_x, max_x, 1000)
y = c + m * x 
```

```python
plt.scatter(X, Y, c='blue', label='Scatter Plot') 
plt.xlabel('year')
plt.ylabel('population')
plt.legend()
plt.show()
```

```python
tot_sumofsq = 0
res_sumofsq = 0
for i in range(n): #n is the total number of values
    y_pred = c + m * X[i]
    tot_sumofsq += (Y[i] - mean_y) ** 2
    res_sumofsq += (Y[i] - y_pred) ** 2
r2 = 1 - (res_sumofsq/tot_sumofsq)
print(r2)
```

```python
plt.plot(x, y, color='red', label='Regression Line')
plt.scatter(X, Y, c='blue', label='Scatter Plot') 
plt.xlabel('year')
plt.ylabel('population')
plt.legend()
plt.show()
```
